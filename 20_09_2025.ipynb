```python
class animal:
    def sound (self):
        print("animal makes a sound")

class dog (animal):
    def sound (self):
        print("bow bow bow bow")

class cat (animal):
    def sound(self):
        print("meow meow meow meow")

class snake(animal) :
    def sound (self):
        print("busssssssssss")

class cow (animal) :
    def sound(self):
        print("ambaaaaaaaaaaaaa")

obj1 = animal()
obj2 = dog()
obj3 = cat()
obj4 = snake()
obj5 = cow()
obj2.sound()
obj1.sound()
obj5.sound()
obj3.sound()
obj4.sound()




```

```python
class Book:
    def __init__(self, page):
        self.page = page

    # Overloading the + operator using __add__
    def __add__(self, other):
        return self.page + other.page

obj1 = Book(200)
obj2 = Book(300)

# Now, this will correctly add the pages
print("Total pages is", obj1 + obj2)
```

```python
from abc import ABC,abstractmethod
class Game(ABC):
    def ff(self):
        print("winner winner fuse dinner")

    def BGMI(self):
        print("i got royal pass diamod")

    def ludo_king(self):
        print("i'm won the game")

class temple_run(Game):
    def BGMI(self):
        print("sagu won")

    def chess(self):
        print("i'm distict player")


obj = temple_run()
obj.BGMI()
obj.ludo_king()
obj.chess()
obj.ff()
```
